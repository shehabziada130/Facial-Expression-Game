```python
import cv2
import numpy as np
from keras.preprocessing.image import img_to_array
from keras.applications.mobilenet import preprocess_input
import time

def capture_and_classify_real_time():
    # Load pre-trained model and other necessary resources
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    cap = cv2.VideoCapture(0)

    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()

        # Convert frame to grayscale for face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Detect faces in the frame
        faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

        for (x, y, w, h) in faces:
            # Crop the detected face
            face_roi = frame[y:y+h, x:x+w]

            # Resize the cropped face for emotion classification
            face_roi = cv2.resize(face_roi, (128, 128), interpolation=cv2.INTER_AREA)

            # Convert BGR to RGB and preprocess the face for classification
            face_rgb = cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB)
            face_array = img_to_array(face_rgb)
            print(face_array.shape)
            face_array = np.expand_dims(face_array, axis=0)
            print(face_array.shape)
            print(face_array)
            face_array = preprocess_input(face_array)
            print(face_array.shape)
            print(face_array)
            
            # Make prediction
            prediction = model.predict(face_array)[0]
            emotion_index = 0 if prediction < 0.5 else 1
            label = emotion_labels_2[emotion_index]

            # Draw rectangle around the detected face and label it with emotion
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

        # Display the frame
        cv2.imshow('Face Classification', frame)

        # Exit loop if 'q' is pressed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the capture
    cap.release()
    cv2.destroyAllWindows()
```
